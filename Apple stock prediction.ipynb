# Data preparation

```python
data.set_index('Date', inplace = True)
```

```python
data.head()
```

```python
x, y = [], []
for i in range(30, len(data)):
  x.append(data.iloc[i-30:i, 1])
  y.append(data.iloc[i,1])
x,y = np.array(x), np.array(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

```python
xtrain = xtrain.reshape(xtrain.shape[0],xtrain.shape[1] , 1)
xtest = xtest.reshape(xtest.shape[0],xtest.shape[1] , 1)
```

# LSTM model

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, GRU, Dropout
```

```python
model= Sequential()
model.add(LSTM(50,return_sequences = True, input_shape = (xtrain.shape[1], 1)))
model.add(LSTM(50,return_sequences = False))   
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss='mean_squared_error')
```

```python
history = model.fit(xtrain, ytrain, batch_size=40, epochs=10)
```

```python
plt.figure(figsize = (15, 8))
sns.lineplot([i for i in range(10)], history.history['loss'])
plt.title('loss of the model')
```

```python
X_test = np.reshape(xtest, (xtest.shape[0],xtest.shape[1],1))
g = model.predict(X_test)
```

```python
x = [i for i in range(2089)]
plt.figure(figsize = (15, 8))
plt.plot(x[:100], ytest[:100], label = 'actual', color = 'red', linewidth = 1.5)
plt.plot(x[:100], g[:100], label = 'predicted', color = 'green', linewidth = 1.5)
plt.title('Plot distribution of Actual and Predicted values')
plt.legend()
```

# predictions using LSTM

```python
X_test = np.reshape(xtest, (xtest.shape[0],xtest.shape[1],1))
g = model.predict(X_test)
g = [i[0] for i in g]
```

```python
x = [i for i in range(2089)]
fig, ax = plt.subplots(2)
sns.lineplot(x[:100], ytest[:100], ax = ax[0])
sns.lineplot(x[:100], g[:100], ax = ax[1])
```

# GRU

```python
regressorGRU = Sequential()
regressorGRU.add(GRU(units=50, return_sequences=True, input_shape=(xtrain.shape[1], 1)))
regressorGRU.add(GRU(50, return_sequences=True))
regressorGRU.add(GRU(50))
regressorGRU.add(Dense(25))
regressorGRU.add(Dense(1))
```

```python
regressorGRU.summary()
```

```python
regressorGRU.compile(optimizer='adam',loss='mean_squared_error')
```

```python
history = regressorGRU.fit(xtrain,ytrain,epochs=10)
```

```python
plt.figure(figsize = (15, 8))
sns.lineplot([i for i in range(10)], history.history['loss'])
plt.title('loss of the model')
```

# predictions using LSTM

```python
X_test = np.reshape(xtest, (xtest.shape[0],xtest.shape[1],1))
g = regressorGRU.predict(X_test)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
random = RandomForestRegressor()
random.fit(xtrain, ytrain)
pred1 = random.predict(xtest)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dec = DecisionTreeRegressor()
dec.fit(xtrain, ytrain)
pred2 = dec.predict(xtest)
```

# Support vector regressor

```python
from sklearn import svm
svr = svm.SVR()
svr.fit(xtrain, ytrain)
pred3 = svr.predict(xtest)
```

# model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
print(r2_score(pred1, ytest))
print(mean_squared_error(pred1, ytest))
```

```python
# decision tree regressor
print(r2_score(pred2, ytest))
print(mean_squared_error(pred2, ytest))
```

```python
# support vector regressor
print(r2_score(pred3, ytest))
print(mean_squared_error(pred3, ytest))
```
